## Marvin
On this [reddit chain](https://www.reddit.com/r/math/comments/cavoz/what_are_some_of_your_favorite_logic_or_tricky/), a post by an unknown redditor stated the following:
>*Marvin gets off work at random times between 3 and 5 pm. His mother lives uptown and girlfriend downtown. He takes the first subway that comes in either direction and eats dinner with who ever he gets delivered to. Marvin's mother complains that Marvin never comes to have dinner with her but he says she has a 50 - 50 chance. Marvin has had dinner with his mother 2 times in the last 20 working days.*

Although the post didn't pose a question, we could imagine a few questions. For example, what is the chance that his mother will have dinner with Marvin on a given day? How often do the trains come? What time will Marvin leave for work on the 21st day? Given timetables of subways, what city does Marvin live in?

Let's start with the first question. If we assume that Marvin indeed leaves work at random time intervals, he has an equal chance of getting off from work at any time, and since in a Bayesian update the likelihoods * priors are normalized, the chance of Marvin leaving from work at a certain time does not affect anything. Since subways come at regular intervals, and if we don't know anything about a specific subway system, the chances that he will get on a train that goes uptown vs downtown is indeed 50/50 (half of the time they go one way, the other half they do the return trip). Therefore, his mother has an equal chance of dining with him as does his girlfriend.

For simplicity, let's assume that trains are always on schedule, so an inbound train comes every couple of minutes and an outbound train comes every couple of minutes. They do not come at the same time and are displaced from each other by a certain time interval. If the interval between the departure of an inbound train and arrival of an outbound train is greater than the interval between the departure of an outbound train and the arrival of an inbound train, a passenger has a higher likelihood of getting on the outbound train.

In this model, only if the trains came every 8 minutes and were displaced 4 minutes from one another would Marvin would have a 50/50 chance of getting on any train.

Suppose trains come every 8 minutes, and outbound trains come 1 minute later than inbound trains. The schedule will look something like this:

**i** <--1min--> **o** <-----7min-----> **i** <--1min--> **o** <-----7min-----> **i** <--1min--> **o** <-----7min----->>>

Marvin has a 7/8 chance of arriving in a time interval between an outbound and inbound train and an only 1/8 chance of arriving between the departure of an inbound train and arrival of an outbound one. Therefore, the probability that he will go downtown to his girlfriend at any given time is 7/8 and not 1/2.

What is the likelihood that Marvin's subway system has such a train time distribution? We know that he has gone on inbound trains 18 times and headed outbound only 2 times. We can model this scenario with a binomial probability because we have 2 cumulatively exhaustive choices:

```python
timeBtwnInOut = 1
timeBtwnOutIn = 7
timeTotal = timeBtwnInOut + timeBtwnOutIn

pOut = timeBtwnInOut / timeTotal # you would catch the outbound train
pIn = 1 - pOut # you would catch the inbound train

# Marvin went inbound 18 times and outbound 2 times
num_inbound = 18
num_outbound = 2
like = pIn ** num_inbound * pOut ** num_outbound
print(like)
```

There is a 0.14 chance that Marvin managed to catch his trains in such a way. Using this forward probability calculation, we can answer a more complicated question: What is the time distribution between the trains? We have two variables, time between an inbound and outbound, and time between the outbound and the next inbound. We look for the most likely pair of times.

We start by enumerating the possible time values. It is highly unlikely for trains to arrive at the exact same time, so the bottom range of time between trains will be 0.1 minutes, while the maximum will be about 20 minutes. Using np.linspace, I created 50 values in this range. Next, we want to zip all of the combinations of the two variable together by using product.

Next, I will create a Suite class which will have a likelihood function that returns the likelihood of the observed data (going uptown twice and downtown 20 times) given one of the hypothesis. The hypothesis in this case will be the possible time distributions.

```python
from itertools import product

timeBtwnInOut = np.linspace(0.1, 10, 100) # inbound left, outbound coming
timeBtwnOutIn = np.linspace(0.1, 10, 100) # outbound left, inbound coming

# All combinations of timeBtwnInOut and timeBtwnOutIn
timeDistributions = product(timeBtwnInOut, timeBtwnOutIn)

class Trains(Suite, Joint):
    '''data: tuple of number of times Marvin goes (outbound, inbound)
    hypo: (timeIO, timeOI) time between in out, and time between out and in
    returns: likelihood that Marvin managed to catch the trains in that pattern given time distribution'''
    def Likelihood(self, data, hypo):
        num_outbound, num_inbound = data
        timeIO, timeOI = hypo
        pO = timeIO / (timeIO + timeOI)
        return pO ** num_outbound * (1 - pO) ** num_inbound

trains = Trains(timeDistributions)
trains.Update((2, 18))
thinkplot.contour(trains)
```

```python
# Distribution of time between inbound and outbound train
thinkplot.Pdf(trains.Marginal(0))
trains.Marginal(0).MaximumLikelihood()
trains.Marginal(0).Mean()
```

```python
# Distribution of time between outbound and inbound train
thinkplot.Pdf(trains.Marginal(1))
trains.Marginal(1).Mean()
```

The graphs of the marginal distributions of our two variables show that the distribution of time between when an inbound train leaves and an outbound train leaves peaks at around 36 seconds (0.6 min) while the distribution of time between when an outbound train departing and an inbound train arriving grows linearly proportional to the duration. These results make sense because the longer the interval between the outbound and inbound train, the more likely Marvin is to arrive during that interval and take an inbound train. The shorter the time between a leaving inbound and arriving outbound train, the more likely Marvin was to have only caught the outbound train twice in 20 days.

The mean values for the two distribution are about 1.26 minutes between an inbound and outbound train and 6.71 minutes between an outbound and inbound trains. Out original random estimation of 1 and 7 minutes, respectively, was actually quite close to what our approximation tells us!

Now let's consider a non linear relationship between train arrivals. If an outbound train just left, the next train could be an inbound train or it could be an outbound train. It is very unlikely that there will be three trains going in the same direction one after the other, so let's just consider that after an inbound train leaves, the next train could be an inbound train or an outbound train.
