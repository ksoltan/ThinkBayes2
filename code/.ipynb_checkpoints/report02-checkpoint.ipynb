## Blue Eyed Children
Problem from <http://math.stackexchange.com/questions/238752/question-on-probability-using-bayes-theorem>

*In a certain day care class, 30% of the children have grey eyes, 50% of them have blue and the other 20%'s eyes are in other colors. One day they play a game together. In the first run, 65% of the grey eye ones, 82% of the blue eyed ones and 50% of the children with other eye color were selected. Now, if a child is selected randomly from the class, and we know that he/she was not in the first game, what is the probability that the child has blue eyes?*

This problem is convenient to solve with the table method. Our goal is to find the probability that the chosen child had a certain color of eyes. We can define our Hypothesis as the chosen child having **blue**, **grey**, or **other** colored eyes. We know the priors to be 50, 30, and 20 percent, respectively, because this is the distribution of eye colors in the class. (Because everything will be normalized, we can write the priors as 5, 3, and 2).

The data given to us is that the child was **not in the first game**. Therefore, we need to calculate the likelihood that the child was not in the game, given that the child has blue eyes, grey eyes, and other colored eyes (aka, under the hypothesis):

**p(D|blue) = 100 - 82 = 18**

**p(D|grey) = 100 - 65 = 35**

**p(D|other) = 100 - 50 = 50**

We look for the likelihood that the child was **NOT** in the game, and we are given the percent of children that were in the game. Putting all of this into a table, we find the posteriors using Bayes' Theorem. We find that there is an 18/59 chance that the child had blue eyes.

## Chess Players
This is a problem by Allen Downey:

*Two identical twins are members of my chess club, but they never show up on the same day; in fact, they strictly 
alternate the days they show up. I can't tell them apart except that one is a better player than the other. Avery beats me 60% of the time, and I beat Blake 70% of the time. If I play one twin on Monday and win, and the other twin on Tuesday and lose, which twin did I play on which day?*

I solved this first using the table method. Our hypothesis will be that I played **Avery (A) on Monday, Blake (B) on Tuesday** and **Blake (B) on Monday, Avery (A) on Tuesday**. Since I have no prior information on who I am playing, there is an equal chance of these two events. The priors are therefore 1 and 1 (to be normalized later).

Next, I have data that I won the game on Monday and lost it on Tuesday. If we take the first hypothesis to be true, winning against Avery is a 40% chance, and losing against Blake is a 30% chance. Because the probability of two independent events happening is their probabilities multiplied, my likelihood is 4 * 3 (also to be normalized later). If I was playing the twins the other way around, I would have a 70% chance of beating Blake, and a 60% chance of losing to Avery, and therefore my likelihood is 7 * 6. In the table, I divided both of the likelihoods by the common factor of 4 to make math simpler.

Putting everything in the table and applying Bayes' Theorem, I have a 2/9 chance that I played Avery on Monday, Blake on Tuesday as compared to a 7/9 chance that I played Blake on Monday and Avery on TUesday.

## Byased Coins
The next few problems are taken from Chapter 4 of Allen Downey's *Think Bayes* book exercises:

*Suppose that instead of observing coin tosses directly, you measure the outcome using an instrument that is not always correct. Specifically, suppose there is a probability y that an actual heads is reported as tails, or actual tails reported as heads.
Write a class that estimates the bias of a coin given a series of outcomes and the value of y.
How does the spread of the posterior distribution depend on y?*

First of all, y would you ask such a question? But puns asyde, this problem is similar to the Euro problem, where given a series of outcomes of coin tosses, we are asked to determine the bias of the coin, or whether its odds of flipping heads or tails is really 50:50, or something closer to 60:40. For this problem, we don't know the exact outcome of the given coin tosses; we are confident about them only y% of the time.

So, to solve this problem, let's define the hypothesis. It is going to be a distribution of the possible coin biasis. We have the freedom to decide what the range of these biasis should be, because it's probably unrealistic to say that the chance of the coin having a bias of 1:99 (or 1 heads for every 99 tails) is equally likely as 50:50 or 99:1. We could use a Beta distribution for the prior and tweak it to have make probabilities closer to 0.5 more likely than others. I chose to make the distribution a little wider, with probabilities greater than 0.2 rising in likelihood until they hit 0.5, and falling in likelihood as they come closer to 0.8.

To make this estimate more accurate, we will use the data of outcomes of coin flips to update the priors. Our likelihoods of the data under the hypothesis will be the likelihood of that particular outcome for some bias of the coin (Hx). Because we are unsure of whether the report of the outcomes was accurate, our likelihoods are, for a bias of x:

**P(Heads|Hx) = y * x** and **P(Tails|Hx) = y * (1-x)**

Let's put this all in a class extending the thinkbayes Suite class.

```python
from sympy import symbols
y = symbols('y')

class Coin(Suite):
    def Likelihood(self, data, hypo):
        likelihood_heads = hypo
        if data == 'H':
            return y * hypo
        else:
            return y * (1 - hypo)
```

I will use a prior that has the beta distribution saying that the likelihood of having a super high bias towards heads or tails super unlikely, and a 50/50 chance to be the most likely.

```python
beta = Beta(10, 10)
thinkplot.Pdf(beta.MakePmf())
```

The Beta class has an update function, but I was not super comfortable using it with the special likelihood function, I created a beta distribution, converted it to a pmf, and used it like a Suite. Although the points do become discrete, for an approximation, they are good enough for a first pass of the problem.

```python
coins = Coin(beta.MakePmf())
coins.UpdateSet(140 *'T' + 110 * 'H')
thinkplot.Pdf(coins)
coins.Mean()
```

An interesting phenomenon happens here: the y variable is cancelled out and has no effect on the data! This is because likelihoods are normalized and both likelihoods are being multiplied by the uncertainty y. So, no matter the certainty with which we record toss outcomes, the average predicted bias of the coin remains 4/9.

## Rhode Problem
*At the 2016 Summer Olympics in the Women's Skeet event, Kim Rhode faced Wei Meng in the bronze medal match. They each hit 15 of 25 skeets, sending the match into sudden death. In the first round, both hit 1 of 2 skeets. In the next two rounds, they each hit 2 skeets. Finally, in the fourth round, Rhode hit 2 and Wei hit 1, so Rhode won the bronze medal, making her the first Summer Olympian to win an individual medal at six consecutive summer games.
But after all that shooting, what is the probability that Rhode is actually a better shooter than Wei? If the same match were held again, what is the probability that Rhode would win?*

*As always, you will have to make some modeling decisions, but one approach is to estimate, for each shooter, the probability of hitting a skeet. Then, to estimate the probability that Rhode is a better shooter, you can draw samples from the two posterior distributions and compare them. To estimate the probability of winning a rematch, you could draw samples from the posterior distributions and simulate a round of 25 shots.*

First, you could estimate what the probability for each of these shooters is of hitting a skeet. Before having any data, we know that they are at Olypmic level of shooting. We could have a distribution of the percentage of skeets they hit. Probably, at Olympic level, it could be that the skeet shooters hit between 50 and 90% of the skeets! If we make a Beta distribution with parameters 3 and 2, it approximately gives such a range:

```python
rhode_distrib = Beta(3, 2)
meng_distrib = Beta(3, 2)
thinkplot.Pdf(rhode_distrib.MakePmf())
```

Now, let's update this prior with the data we are given. The next couple of cells are sequential updates for each of the rounds that they play. The first parameter of the tuple is the number of hits and the second parameter is the number of misses for that round.

```python
# Round 1
rhode_distrib.Update((15, 10))
meng_distrib.Update((15,10))
thinkplot.Pdf(rhode_distrib.MakePmf())
thinkplot.Pdf(meng_distrib.MakePmf())
```

```python
# Round 2
rhode_distrib.Update((1, 1))
meng_distrib.Update((1,1))
thinkplot.Pdf(rhode_distrib.MakePmf())
thinkplot.Pdf(meng_distrib.MakePmf())
```

```python
# Round 3
rhode_distrib.Update((2, 0))
meng_distrib.Update((1,1))
thinkplot.Pdf(rhode_distrib.MakePmf())
thinkplot.Pdf(meng_distrib.MakePmf())
```

Before Round 3, the distributions are exactly the same because the update data is the same. Only in round 3 does Meng's one miss make the distributions different.

Now that we have the distribution of how accurate the shooters can be, we can figure out who is better by randomly sampling the distribution, or randomly sampling percentage targets each can hit. In this sample, if Rhode has more higher probabilities than Meng, she is better, and vice versa:

```python
rhode_sample = rhode_distrib.MakeCdf(100001).Sample(10000)
meng_sample = meng_distrib.MakeCdf(100001).Sample(10000)
np.mean(rhode_sample > meng_sample)
```

According to this approximation, Rhode is in 60% of cases better than Meng. Now what is the probability that Rhode will win a rematch against Meng?

Both women are pretty similar in performance, and this is a game of chance as much as skill because in competitions, who doesn't lose?

We know the distribution of the probability that each one will hit a target, or if you'd like, the likelihood that this is the probability with which she will hit a target on a given shoot. For example, there is a about a 5% chance that Rhode will shoot with 60% accuracy (that was an estimation from the distribution above on the graph).

So, we can proceed in two ways. We could simulate a rematch by randomly picking a probability with which each one shoots, randomly "flipping" the coin with that probability and recording whether they miss or not. Then, after running such test a couple of times, take their average and figure out who hits the highest percentage of tagets in all.

Another way we could do it is go through each of the possible probabilities with which they could shoot, and get a posterior with a distribution of the probabilities of winning.
```python
import random
'''Models a rematch between player 1 and 2 by using random samples of probabilities of 
them hitting a target, shooting, and averaging the results.
sample1: list of samples of probabilities for player 1
sample2: list samples of probabilities for player 2
return: tuple of probability player 1 winning and player 2 winning'''
def rematch(sample1, sample2):
    rhode_hits = 0
    meng_hits = 0
    for p1, p2 in zip(sample1, sample2):
        for i in range(25):
            if shoot(p1):
                rhode_hits += 1
            if shoot(p2):
                meng_hits += 1
    return (rhode_hits/(25*len(sample1)), meng_hits/(25*len(sample2)))

'''Given the probability of shooting a skeet, return True p percent of the time
Because ramdon.random return number between 0 and 1, there is a probability p you 
will get something between 0 and p, and probability 1-p that you get something between 
p and 1. That way, if you land between 0 and p, that represents landing there p percent of the time'''

def shoot(p):
    return random.random() < p

rematch(rhode_sample, meng_sample)
```

With this method, Rhode has about a 61% chance of winning a rematch. However, this estimate will keep on changing a little based on the random selection of the data.

## Books
*I picked up a book from a free shelf at the library. The front and back pages are torn out, and I do not know when it was written. Upon reading the book, I found 10 slang words. If every year starting from 1900 a new slang word is added to the lexicon, when was the book written?*

There are many choices for when the book was written, from 1900 onward to 2016. Because this distribution is discrete, we should use a pmf as the prior distribution. For now, let's assume that it is a uniform distribution.

The answer we need to answer with our likelihood calculation is what is the probability that a book from year x has 10 slang words? If the year is 1900, there is 1 slang word in the vocabulary, and there is 0 percent chance that there are 10 slang words. If the year is 1930, there are 30 slang words in the vocabulary. What is the probability of choosing 10 slang words from 30? Each word has a 1/30 chance of getting chosen. We can use a binomial likelihood function to determine this. We have 30 words (n), 10 we want to choose (k), and each word has a 1/30 chance of being chosen (p).

I used an extension of Suite to encode this information using the thinkbayes framework:

```python
from scipy.special import binom
import scipy.stats as stats

class Books(Suite):
    '''data: number of slang words in text
       hypo: year
       returns: likelihood that there are that many slang words in a book published in that year'''
    
    def Likelihood(self, data, hypo):
        num_words = hypo - 1900
        if data > num_words:
            return 0
        
        p = 1 / num_words
        return stats.binom.pmf(data, num_words, p)

books = Books(range(1900, 2016))
books.Update(10)
thinkplot.Pdf(books)
books.Mean()
```

The result we get makes sense: the more slang words in a language, the more combinations of these slang words you can make, and therefore the more likely the book is recent. If we take the mean of the posteriors, we get about 1976 as the book's possible publication year.
